```python
for i in df.index:
    if df['drct'][i].isalpha() == True:
        df['drct'][i] = np.nan

    print(df['sknt'][i], df['drct'][i])
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df['sknt'] = df['sknt'].astype(float)
```

```python
df['drct'] = df['drct'].astype(float)
```

```python
# Create a WindroseAxes
ax = WindroseAxes.from_ax()
```

```python
# Plot the windrose graph
ax.bar(df['drct'], df['sknt'], normed=True, opening=0.8, edgecolor='white')
```
